```python
s1 = "00000010000000001000000"#use Kappa coefficient K measures agreement 
s2 = "00000001000000010000000"
s3 = "00010000000000000001000"
print(nltk.windowdiff(s1, s1, 3))#the window had a size of 3. 
'''
#（1）设置的窗口从字符串最左边开始
#（2）计算窗口内两个字符串边界的数的差值，差值的绝对值累加（weighted值决定是加1还是差值）
#（3）窗口向右移动，跳转2
#（4）如果到达最右端，跳出循环
#（5）边界的数的差值和/(字符串长度-窗口大小+1），窗口出现的次数
'''

print(nltk.windowdiff(s1, s2, 3))

print(nltk.windowdiff(s2, s3, 3))
```

```python
import csv
import nltk

#CSV格式处理

lexicon = csv.reader(open('dict.csv',encoding='utf-8-sig'))
'''
用csv处理的包读取文件，如果encoding=‘utf-8’,输出中间处理的结果，第一个字符串前面会出现\ufeff

encoding=‘utf-8-sig’


在编码/编写时，UTF-8编码的BOM（ Byte Order Mark ）会被当作前缀以三个字节写入。

如果前三个字节是UTF-8编码BOM，则在解码/读取时，这些字节将被跳过。
'''
pairs = [(lexeme, defn) for (lexeme, _, _, defn) in lexicon]
print(pairs)
lexemes, defns = zip(*pairs)

defn_words = set(w for defn in defns for w in defn.split())#[1]



idx = nltk.Index((defn_word, lexeme)  #建立词语和词意的索引
                 for (lexeme, defn) in pairs#对文本的解释分词和
                 for defn_word in nltk.word_tokenize(defn)  
                 if len(defn_word) > 3) #舍弃短的短语

with open("dict.idx", "w") as idx_file:
     for word in sorted(idx):
         idx_words = ', '.join(idx[word]) 
         print(idx_words)        
         idx_line = "{}: {}".format(word, idx_words)
         print( idx_line)
         print(idx_line, file=idx_file)
```

```python
from collections import Counter


speaker_seq = [s.text for s in merchant.findall('ACT/SCENE/SPEECH/SPEAKER')]
#findall(‘ACT/SCENE/SPEECH/SPEAKER’)找到路径下所有的speaker，
speaker_freq = Counter(speaker_seq)
#用collections的Counter统计每个speaker出现的次数，选取top5
top5 = speaker_freq.most_common(5)
top5
```

```python
from collections import defaultdict
abbreviate = defaultdict(lambda: 'OTH')
for speaker, _ in top5:
     abbreviate[speaker] = speaker[:4]

speaker_seq2 = [abbreviate[speaker] for speaker in speaker_seq]

print(nltk.bigrams(speaker_seq2))
cfd = nltk.ConditionalFreqDist(nltk.bigrams(speaker_seq2))
cfd.tabulate()
```

```python
import nltk
from xml.etree.ElementTree import SubElement
import re


'''
将辅音和元音的字符串映射到相应的CV 序列，即：kakapua 将映射到CVCVCVV。这种映射有四个步骤。首先，
1，将字符串转换为小写，那么将所有非字母字符[^a-z]用下划   代替。下一步，
2，将所有元音替换为V。
3，所有不是V 或下划线的必定是一个辅音，所以我们将它替换   为C
   扫描词汇，在每个lx 字段后面添加一个新的cv 字段。
'''


def cv(s):
    s = s.lower()#将字符串转换为小写
    s = re.sub(r'[^a-z]',     r'_', s)#将所有非字母字符[^a-z]用下划代替
    s = re.sub(r'[aeiou]',    r'V', s)#将所有元音替换为V
    s = re.sub(r'[^V_]',      r'C', s)#所有不是V 或下划线的必定是一个辅音，所以我们将它替换为C
    return (s)

def add_cv_field(entry):
#扫描词汇，在每个lx 字段后面添加一个新的cv 字段。
    for field in entry:
        if field.tag == 'lx':
            cv_field = SubElement(entry, 'cv')
            cv_field.text = cv(field.text)
lexicon = toolbox.xml('rotokas.dic')
add_cv_field(lexicon[53])
print(nltk.toolbox.to_sfm_string(lexicon[53]))
```

```python
'''
文法使用我们在第8 章看到的CFG 的格式。
这样的文法模型隐含Toolbox 条目的嵌套结构，
建立一个树状结构，树的叶子是单独的字段名。
我们遍历条目并报告它们与文法的一致性，
。那些被文法接受的在前面加一个“+，那些被文法接受的在前面加一个“-”。
在开发这样一个文法的过程中，它可以帮助过滤掉一些标签
'''

grammar = nltk.CFG.fromstring('''
  S -> Head PS Glosses Comment Date Sem_Field Examples
  Head -> Lexeme Root
  Lexeme -> "lx"
  Root -> "rt" |
  PS -> "ps"
  Glosses -> Gloss Glosses |
  Gloss -> "ge" | "tkp" | "eng"
  Date -> "dt"
  Sem_Field -> "sf"
  Examples -> Example Ex_Pidgin Ex_English Examples |
  Example -> "ex"
  Ex_Pidgin -> "xp"
  Ex_English -> "xe"
  Comment -> "cmt" | "nt" |
  ''')

def validate_lexicon(grammar, lexicon, ignored_tags):
    rd_parser = nltk.RecursiveDescentParser(grammar)
    print(rd_parser)
    for entry in lexicon:
        marker_list = [field.tag for field in entry if field.tag not in ignored_tags]
        if list(rd_parser.parse(marker_list)):
            print("+", ':'.join(marker_list)) 
        else:
            print("-", ':'.join(marker_list)) 

lexicon = toolbox.xml('rotokas.dic')[10:20]
ignored_tags = ['arg', 'dcsv', 'pt', 'vx'] 
validate_lexicon(grammar, lexicon, ignored_tags)
```

```python
#语言记录的示例
mappings = [('ph', 'f'), ('ght', 't'), ('^kn', 'n'), ('qu', 'kw'),
          ('[aeiou]+', 'a'), (r'(.)\1', r'\1')]
#易混淆字母序列和发音的映射
def signature(word):
    for patt, repl in mappings:
           word = re.sub(patt, repl, word)
    pieces = re.findall('[^aeiou]+', word)
    return ''.join(char for piece in pieces for char in sorted(piece))[:8]


print(signature('illefent'))

print(signature('ebsekwieous'))

print(signature('nuculerr'))
```
